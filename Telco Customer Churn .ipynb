```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style
import seaborn as sns
```

## Exploratory Data Analysis(EDA)


```python
df = pd.read_csv('Customer Churn.csv')
```

## 2nd Step Replace Blanks with 0 as tenure is 0 and total charges are recorded and change the type from object to float 

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "NO"

        
```

```python
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
df.head(60)
```

```python
ax = sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title("count of Customer by Churns")
plt.show()
```

```python
gb = df.groupby('Churn').agg({'Churn':'count'})
```

```python
gb
```

```python
plt.figure(figsize = (5,5))
plt.pie(gb['Churn'], labels = gb.index, autopct= "%1.2f%%")
plt.title("Percentage of Churn Customers")
plt.show()
```

## It included that gender has nothing to do with churn out 

```python
gb1 = df.groupby("SeniorCitizen")['Churn'].value_counts(normalize = True).unstack() *100
fig, ax = plt.subplots(figsize=(6,6))
gb1.plot(kind = 'bar', stacked = True, ax=ax, color = ['#1f77b4', '#ff7f0e'])
for p in ax.patches:
    width,height = p.get_width(), p.get_height()
    x,y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.2f}%' , ha = 'center', va= 'center')
plt.xlabel("SeniorCitizen ")
plt.ylabel("percentage %")
plt.title("Churn by Senior citizen(Stacked bar Chart)")
plt.legend(title = "Churn", bbox_to_anchor=(0.9,0.9))
plt.xticks(rotation = 0)
plt.show()
```

```python
df.columns.values
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots
num_cols = 3  # Number of columns in the subplot grid
num_rows = (len(columns) + num_cols - 1) // num_cols  # Calculate rows needed

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10),)
axes = axes.flatten()  # Flatten the 2D array of axes

# Loop through the columns and create a countplot in each subplot
for i, column in enumerate(columns):
    sns.countplot(x=column, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(column)
    axes[i].set_ylabel('Count')
    axes[i].set_xlabel('')  # Optionally remove x-labels for better aesthetics

# Hide any empty subplots
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

## The data suggests that services like Fiber Optic Internet and lack of Tech Support or Online Security have higher churn rates. Services without internet or phone services tend to show less churn impact overall.

```python
plt.figure(figsize=(9,4))
ax2 = sns.countplot(x = "PaymentMethod", data = df, hue = "Churn")
ax2.bar_label(ax2.containers[0])
ax2.bar_label(ax2.containers[1])
plt.xticks(rotation = 45)
plt.title("Churn vCustomer by Payment Method")
plt.show()
```
